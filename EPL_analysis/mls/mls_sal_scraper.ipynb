```python
import re

import requests
from bs4 import BeautifulSoup
import os

from scipy.ndimage import value_indices
```

```python
url='https://mlsplayers.org/resources/salary-guide'
req=requests.get(url)
site=req.text
soup=BeautifulSoup(site,'html.parser')
```

```python
salsa=soup.find("div", class_="salaryGuides bodyContent")
```

```python
str(salsa)
```

```python
seaz = re.findall(r'http://s3.amazonaws.com/mlspa.*?Salary Report', str(salsa))
```

```python
first=seaz[0]
```

```python
first_url=first.split("\"")[0]
```

```python
first_s=first.split("\"")[-1][2:6]
```

```python
pdf_dict={}
```

```python
pdf_dict.update({first_s:first_url})
```

```python
pdf_dict
```

```python
for i in seaz:
    first_url=i.split("\"")[0]
    first_s=i.split("\"")[-1][2:6]
    pdf_dict.update({first_s:first_url})
```

```python
pdf_dict
```

```python
for k, v in pdf_dict.items():
    print(f"{v}")
```

```python
if not os.path.exists('mls_salaries'):
    os.mkdir('mls_salaries')
for key, value in pdf_dict.items():
    req=requests.get(f"{value}")
    url_cont= req.content
    file_csv=open('mls_salaries/'+f"{key}"+'season.csv','w')
    file_csv.write(str(url_cont))
    file_csv.close()
    
```
