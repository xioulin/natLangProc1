```python
import nltk
import gensim
from gensim.models import Word2Vec  
from nltk import word_tokenize,sent_tokenize
from nltk.probability import FreqDist
```

```python
nltk.download('punkt')
nltk.download('gutenberg')
```

```python
moby_dick = open("PequodProject/mobyDick.txt")
with moby_dick as f:
    book= f.readlines()
moby_dick.close()
```

```python
book = str(book)
```

```python
data = []
for sent in sent_tokenize(book):
    temp = []
    for word in word_tokenize(sent):
        temp.append(word.lower())
    data.append(temp)
```

```python
#creating a skip gram model
model1= gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5, sg=1)
```

```python
model1.wv.similarity('cannibal','queequeg')
```

```python
model1.wv.most_similar('platonic')
```

```python
model1.wv.similarity('platonic','cannibal')
```

```python
model1.wv.most_similar('mad')
```
