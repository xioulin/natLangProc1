url is collected works of shakespeare, pull from url and render it as a text

```python
url='https://www.gutenberg.org/cache/epub/100/pg100-images.html'
req=requests.get(url)
collected_works=req.text
```

```python
html=re.findall('<.*?>',play)
for i in html:
    play=play.replace(i," ")
```

```python
for i in thingy:
    play = play.replace(i, " ")
```

```python
characters=re.findall("Dramatis.*?SCENE",play)[0][17:]
```

```python
characters
```

```python
descr=re.findall("[A-Z].+? [A-Z][a-z].",characters)
```

```python
descr
```

```python
char_dict={}
for i in descr:
    f_name= re.search(r'[A-Z]{3}.*?\b',i).group()
    char_d= re.search(r"\s.*",i).group()
    # row = {'name':f_name,'descr':char_d}
    # print(f_name,char_d)
    char_dict.update({f_name:char_d})
```
