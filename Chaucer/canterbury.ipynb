```python
import nltk
import requests as r
```

```python
from nltk.corpus import gutenberg
```

```python
r = r.get(r'https://www.gutenberg.org/cache/epub/2383/pg2383.txt')
```

```python
canter=r.text
```

```python
for char in ['\r','\t']:
    canter= canter.replace(char,' ')
```

```python
endpoint= canter.index('Notes to the Prologue to the Parson’s Tale')
```

```python
chaucer=canter[:endpoint]
```

```python
prologue=chaucer.index('Inspired hath in every hol')
```

```python
chaucer=chaucer[prologue-455:]
```

```python
chaucer=chaucer[175:]
```

```python
chaucer= chaucer.replace('\n','')
```

```python
chaucer=chaucer[10:]
```

```python
chaucer[:34500]
```

```python
chaucer.index('Theseus')
```

```python
prologue=chaucer[:51035]
```

```python
from nltk import word_tokenize,sent_tokenize,pos_tag
```

```python
toked_prol=word_tokenize(prologue)
```

```python
pos_tag(toked_prol)
```

```python
import re
```

```python
prologue
```

```python
notes=prologue.index('Notes to the Prologue')
```

```python
prologue=prologue[:notes]
```

```python
re.findall(r'\*[a-z][A-Z]*',prologue)
```

```python
numberfication= re.findall(r'<[0-9]>|<[0-9][0-9]>',prologue)
```

```python
for i in numberfication:
    prologue=prologue.replace(i,'')
```

```python
prologue
```

```python
asterisks=re.findall(r'\*[a-z]*\b',prologue)
```

```python
for i in asterisks:
    prologue=prologue.replace(i,'')
```

```python
prologue
```

```python
prologue_clean=open('corporas/prologue_clean.txt','w')
```

```python
prologue_clean.write(prologue)
```
