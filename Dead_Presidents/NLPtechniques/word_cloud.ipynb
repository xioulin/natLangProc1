Jupyter notebook looking at Lincoln's First Inaugural Address, and two speeches by southern secessionists, one by the eventual vice president of the confederacy Stephens and another by a leading Georgia secessionist Benning. There really isn't all that much interesting from making word clouds for either of these.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
linc_1st=open('../addressesraw/1861-Lincoln.txt','r')
```

```python
linc1=linc_1st.read()
```

```python
linc1.replace('\n','')
```

```python
wc=WordCloud(height=200,width=200).generate(linc1)
```

```python
plt.imshow(wc,interpolation='bilinear')
```

```python
benning= open('../benning_speech.txt','r')
benning= benning.read()
```

```python
wc2= WordCloud(height=200,width=200).generate(benning)
plt.axis('off')
plt.imshow(wc2,interpolation='bilinear')
```

```python
cornerstone=open('../cornerstone_speech.txt','r')
c_stone=cornerstone.read()
c_stone= c_stone.replace('\n','')
```

```python
def stringToWC(string):
    string= string.replace('\n','')
    wc= WordCloud().generate(string)
    plt.axis('off')
    plt.imshow(wc,interpolation='bilinear')
```

```python
wc3=WordCloud(height=200,width=200).generate(c_stone)
plt.axis('off')
plt.imshow(wc3,interpolation='bilinear')
```
