```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk import word_tokenize
from nltk.probability import FreqDist
from useful_NLP_functions import filterStopandPunctList
```

```python
def stringToWC(string):
    string= string.replace('\n','')
    wc= WordCloud(height=200,width=200).generate(string)
    plt.axis('off')
    plt.imshow(wc,interpolation='bilinear')
```

```python
file= open('addressesraw/1861-Lincoln.txt','r')
```

```python
first= file.read()
```

```python
file.close()
```

```python
wc= WordCloud(height=200,width=200).generate(first)
```

```python
plt.imshow(wc,interpolation='bilinear')
```

```python
wc.to_file("wordclouds/linc_first.png")
```

```python
first=first.replace('\n','')
```

```python
w= word_tokenize(first)
```

```python
clean= filterStopandPunctList(w)
```

```python
freq= FreqDist(clean)
```

```python
freq.plot(10)
```
