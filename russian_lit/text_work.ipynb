```python
notes = open('notes_from_underground_cleaner.txt','rb')
```

```python
notes = notes.read()
```

```python
import re
```

```python
notes=str(notes)
```

```python
weird=re.findall(r'\\.*\s',notes)
```

```python
from nltk import word_tokenize
```

```python
toked= word_tokenize(notes)
```

```python
last500=toked[-500:]
```

```python
last500
```

```python
weird_stuff=[]
for i in last500:
    if '\\' in i:
        weird_stuff.append(i)
        print(i)
```

```python
wordish= weird_stuff[0]
```

```python
wordish
```

```python
match=re.findall(r'\\[a-z]*[0-9]*',wordish)
```

```python
match
```

```python
all_the_weird_stuff= re.findall(r'\\[a-z]*[0-9]',notes)
```

```python
all_the_weird_stuff
```

```python
for i in toked:
    for j in all_the_weird_stuff:
        if j in i:
            i.replace(j,'')
```

```python
for i in all_the_weird_stuff:
    notes= notes.replace(i,'')
```

```python
notes
```

```python
cleaner = open('notes_cleaner','w')
cleaner.write(notes)
cleaner.close
```
